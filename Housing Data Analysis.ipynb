# Modelling

```python
# Importing Dependendncies
from  sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = df_Metadata[['Rooms', 'Type',
       'Distance', 'Bedroom2', 'Bathroom', 'Car', 'Landsize',
       'BuildingArea', 'YearBuilt', 'Price']]
df.dtypes
```

```python
from sklearn.preprocessing import StandardScaler
df_standardize = pd.DataFrame(StandardScaler().fit_transform(df),columns=['Rooms', 'Type',
       'Distance', 'Bedroom2', 'Bathroom', 'Car', 'Landsize',
       'BuildingArea', 'YearBuilt','Price'])
df_standardize.head()
```

```python
x_feature = df_standardize[['Rooms', 'Type',
       'Distance', 'Bedroom2', 'Bathroom', 'Car', 'Landsize',
       'BuildingArea', 'YearBuilt']]
y_label = df_standardize['Price']
```

```python
x_train,x_test,y_train_labels,y_test_labels = train_test_split(x_feature,y_label, test_size = .2,random_state=2) 
```

```python
logreg = LinearRegression()
```

```python
logreg.fit(x_train,y_train_labels)
Y_pred = logreg.predict(x_test)
```

```python
preidct = logreg.predict(x_test)
preidct
```

```python
logreg.score(x_test,y_test_labels)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rms = sqrt(mean_squared_error(y_test_labels, Y_pred))
rms
```
